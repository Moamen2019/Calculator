```python
import math
##Define is float function to check number
def isfloat(num):
  try:
      float(num)
      return True
  except ValueError:
      return False

##Define addition function taking num1 & num2 as arguments
def add(num1,num2):
    result=num1+num2
    return result

  ##Define subtraction function taking num1 & num2 as arguments
def subtract(num1,num2):
    result=num1-num2
    return result

  ##Define multiplication function taking num1 & num2 as arguments
def multiply(num1,num2):
    result=num1*num2
    return result

  ##Define division function taking num1 & num2 as arguments
def divide(num1,num2):
    result=num1/num2
    return result

  ##Define factorial function taking num1 as an argument:
def factorial(num1):
  """This is a recursive function
  to find the factorial of an integer"""

  if num1 == 1:
      return 1
  else:
      return (num1 * factorial(num1-1))

##Prompt the user to select an operation from (+,-,*,/,!) and checks it
##If any of them is entered , the process is repeated till a valid operator is selected


while True:
  operator=input('Enter an operator(+,-,*,/,!,sqrt,power,sin,cos,tan,log,ln) are only allowed:\n')

  if operator=='+':
    print('Add operator is selected\n')
    number1=input('Enter num 1 :')
    while isfloat(number1)==False:
      number1=input('Please enter a valid num 1 :')
      
    else:
      number1=float(number1)

    number2=input('Enter num 2 :')
    while isfloat(number2)==False:
      number2=input('Please enter a valid num 2 :')
      
    else:
      number2=float(number2)

    add_result=add(number1,number2)
    print('Addition result is :')
    print(add_result)
    temp=input('Do you want to make another calculation? Y or N\n')
    if temp.upper()=="N":
      print("Good Bye Then!")
      break
    else:
      continue

  elif operator=='-':
    print('subtract operator is selected\n')
    number1=input('Enter num 1 :')
    while isfloat(number1)==False:
      number1=input('Please enter a valid num 1 :')
    else:
      number1=float(number1)
      
    number2=input('Enter num 2 :')
    while isfloat(number2)==False:
      number2=input('Please enter a valid num 2 :')
    else:
      number2=float(number2)

    subtract_result=subtract(number1,number2)
    print('Subtraction result is :')
    print(subtract_result)
    temp=input('Do you want to make another calculation? Y or N\n')
    if temp.upper()=="N":
      print("Good Bye Then!")
      break
    else:
      continue

    

  elif operator=='*':
    print('multiply operator is selected\n')
    number1=input('Enter num 1 :')
    while isfloat(number1)==False:
      number1=input('Please enter a valid num 1 :')
    else:
      number1=float(number1)
      
    number2=input('Enter num 2 :')
    while isfloat(number2)==False:
      number2=input('Please enter a valid num 2 :')
    else:
      number2=float(number2)
    multiply_result=multiply(number1,number2)
    print('Multiplication result is :')
    print(multiply_result)
    temp=input('Do you want to make another calculation? Y or N\n')
    if temp.upper()=="N":
      print("Good Bye Then!")
      break
    else:
      continue

  elif operator=='/':
    print('divide operator is selected\n')
    number1=input('Enter num 1 :')
    while isfloat(number1)==False:
      number1=input('Please enter a valid num 1 :')
    else:
      number1=float(number1)
      
    number2=input('Enter num 2 :')
    while isfloat(number2)==False:
      number2=input('Please enter a valid num 2 :')
    else:
      number2=float(number2)
    divide_result=divide(number1,number2)
    print('Division result is :')
    print(divide_result)
    temp=input('Do you want to make another calculation? Y or N\n')
    if temp.upper()=="N":
      print("Good Bye Then!")
      break
    else:
      continue


  elif operator=='!':
    print('factorial operator is selected\n')
    number1=input('Enter num 1 :')
    while isfloat(number1)==False:
      number1=input('Please enter a valid num 1 :')
    else:
      number1=float(number1)
    
    factorial_result=factorial(number1)
    print('Factorial of '+str(number1)+' = '+ str(factorial_result))
    temp=input('Do you want to make another calculation? Y or N\n')
    if temp.upper()=="N":
      print("Good Bye Then!")
      break
    else:
      continue

  elif operator=='sqrt':
    print('square root operator is selected\n')
    number1=input('Enter num 1 :')
    while isfloat(number1)==False:
      number1=input('Please enter a valid num 1 :')
    else:
      number1=float(number1)

    sqrt_result=math.sqrt(number1)
    print('Square Root of '+str(number1)+' is :')
    print(sqrt_result)
    temp=input('Do you want to make another calculation? Y or N\n')
    if temp.upper()=="N":
      print("Good Bye Then!")
      break
    else:
      continue

  elif operator=='power':
    print('Power operator is selected\n')
    number1=input('Enter num 1 :')
    while isfloat(number1)==False:
      number1=input('Please enter a valid num 1 :')
    else:
      number1=float(number1)

    number2=input('Enter num 2 :')
    while isfloat(number2)==False:
      number2=input('Please enter a valid num 2 :')
    else:
      number2=float(number2)

    power_result=pow(number1,number2)
    print('Power result is :')
    print(power_result)
    temp=input('Do you want to make another calculation? Y or N\n')
    if temp.upper()=="N":
      print("Good Bye Then!")
      break
    else:
      continue

  elif operator=='sin':
    print('Sin operator is selected\n')
    number1=input('Enter num 1 :')
    while isfloat(number1)==False:
      number1=input('Please enter a valid num 1 :')
    else:
      number1=float(number1)

    number1=math.radians(number1)
    sin_result=math.sin(number1)
    print('Sin of angle '+str(number1)+' in radians is :')
    print(sin_result)
  
  elif operator=='cos':
    print('Cos operator is selected\n')
    number1=input('Enter num 1 :')
    while isfloat(number1)==False:
      number1=input('Please enter a valid num 1 :')
    else:
      number1=float(number1)

    number1=math.radians(number1)
    cos_result=math.cos(number1)
    print('Cos of angle '+str(number1)+' in radians is :')
    print(cos_result)
    temp=input('Do you want to make another calculation? Y or N\n')
    if temp.upper()=="N":
      print("Good Bye Then!")
      break
    else:
      continue

  elif operator=='tan':
    print('tan operator is selected\n')
    number1=input('Enter num 1 :')
    while isfloat(number1)==False:
      number1=input('Please enter a valid num 1 :')
    else:
      number1=float(number1)

    number1=math.radians(number1)
    tan_result=math.tan(number1)
    print('Tan of angle '+str(number1)+' in radians is :')
    print(tan_result)
    temp=input('Do you want to make another calculation? Y or N\n')
    if temp.upper()=="N":
      print("Good Bye Then!")
      break
    else:
      continue

  elif operator=='log':
    print('Log operator is selected\n')
    number1=input('Enter num 1 :')
    while isfloat(number1)==False:
      number1=input('Please enter a valid num 1 :')
    else:
      number1=float(number1)

    base=input('Enter base :')
    while isfloat(base)==False:
      base=input('Please enter a valid base :')
    else:
      base=float(base)

    log_result=math.log(number1,base)
    print('Log result for '+str(number1)+ ' to base '+str(base)+' is :')
    print(log_result)
    temp=input('Do you want to make another calculation? Y or N\n')
    if temp.upper()=="N":
      print("Good Bye Then!")
      break
    else:
      continue

  elif operator=='ln':
    print('Log operator is selected\n')
    number1=input('Enter num 1 :')
    while isfloat(number1)==False:
      number1=input('Please enter a valid num 1 :')
    else:
      number1=float(number1)

    ln_result=math.log(number1)
    print('Ln result for '+str(number1)+' is :')
    print(ln_result)
    temp=input('Do you want to make another calculation? Y or N\n')
    if temp.upper()=="N":
      print("Good Bye Then!")
      break
    else:
      continue

  else:
    print('Please enter a valid operator from the above list')
    continue
```
